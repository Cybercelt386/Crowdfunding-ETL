```python
# creat lists for each information to extract
backer_ids = []
cf_ids = []
names = []
emails = []
```

```python
p = r'\s\d{2,4}'
backer_df[backer_df['backer_info'].str.contains(p)]
```

```python
pattern = re.compile(p)
matches = pattern.findall(str(backer_df['backer_info'].to_list()))
for match in matches:
       # print(f'cf_id: {match}')
        cf_ids.append(int(match))
        
print(len(cf_ids))
```

```python
print(backer_df['backer_info'])
```

```python
cf_df = pd.DataFrame(cf_ids, columns=["cf_id"])
cf_df
```

```python
p = r'\w{2}\d{3}'
backer_df[backer_df['backer_info'].str.contains(p)]
```

```python
pattern = re.compile(p)
matches = pattern.findall(str(backer_df['backer_info'].to_list()))
for match in matches:
    # print(match)
    backer_ids.append(match)

print(len(backer_ids))
```

```python
bkr_ids_df = pd.DataFrame(backer_ids, columns= ["backer_id"])
bkr_ids_df
```

```python
p = r'\w{2,}@\w{2,}\.com'
backer_df[backer_df['backer_info'].str.contains(p)]
```

```python
pattern = re.compile(p)
matches = pattern.findall(str(backer_df['backer_info'].to_list()))
for match in matches:
    # print(match)
    emails.append(match)
    
print(len(emails))
```

```python
email_df = pd.DataFrame(emails, columns = ["email"])
email_df
```

```python
p = r'\"?[A-Z]\w{1,}\s[A-Z]\w{1,}'
backer_df[backer_df['backer_info'].str.contains(p)]
```

```python
pattern = re.compile(p)
matches = pattern.findall(str(backer_df['backer_info'].to_list()))
for match in matches:
    # print(match)
    names.append(match)
    
print(len(names))
```

```python
names_df = pd.DataFrame(names)
names_df
```

```python
new_info = {"Backer_ID": backer_ids, "CF_ID": cf_ids, "Name": names, "Email": emails }
```

```python
df = pd.DataFrame(new_info, columns=['Backer_ID', 'CF_ID', 'Name', 'Email'])
df[["First_Name","Last_Name"]] = df["Name"].str.split(' ', n=1, expand=True)
df = df.drop(columns=["Name"])
df
```
